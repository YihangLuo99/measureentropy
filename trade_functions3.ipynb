```python
def get_trade_info(lst, index):
    n = len(lst)
    if not (0 <= index < n):
        raise IndexError("Index out of range")

    # Step 1: Calculate immediate neighbors (Left/Right in a circle)
    idx_prev = (index - 1) % n
    idx_next = (index + 1) % n

    # Step 2: Calculate neighbors' neighbors (Two steps away)
    idx_prev_prev = (idx_prev - 1) % n
    idx_next_next = (idx_next + 1) % n

    # Step 3: Grouping
    # n1: The "previous" side chain
    n1 = [idx_prev, idx_prev_prev]
    # n2: The "next" side chain
    n2 = [idx_next, idx_next_next]
    
    # n3: All unique excluded neighbors
    excluded_set = {idx_prev, idx_prev_prev, idx_next, idx_next_next}
    n3 = list(excluded_set)

    # Step 4: Generate tradable list
    # Exclude self and the 4 neighbors in n3
    all_excluded = excluded_set | {index}
    tradable_list = [i for i in range(n) if i not in all_excluded]

    return n1, n2, n3, tradable_list
```

```python
# define a function to sample
def random_pick(g, G, a):
    # Randomly select a number within the range [g-mg, g+mg]
    delta_g=G*a
    new_g = random.uniform(g - delta_g, g + delta_g)

    # Apply the reflex logic
    if new_g < 0:
        new_g = - new_g
    elif new_g > G:
        new_g = 2*G - new_g

    return new_g
```

```python
def conditional_weight(g, m, g_new, m_new, alpha, eta):
    """
    w(g', m' | g, m) = (g + g')^(alpha-1) * (m + m')^(eta-1)
    """
    return (g + g_new)**(alpha - 1) * (m + m_new)**(eta - 1)
```

```python
import random

def sample_feasible_proposal(g, m, G, M, mu1, mu2, max_tries=1000):
    """
    Sample (g', m') such that
    mu1 <= (m' - m) / (g - g') <= mu2
    """
    for _ in range(max_tries):
        g_new = random.random() * G
        dg = g - g_new
        if abs(dg) < 1e-10:
            continue

        if dg > 0:
            dm_min = mu1 * dg
            dm_max = mu2 * dg
        else:
            dm_min = mu2 * dg
            dm_max = mu1 * dg

        m_low = max(0.0, m + dm_min)
        m_high = min(M,   m + dm_max)

        if m_low < m_high:
            m_new = m_low + random.random() * (m_high - m_low)
            return g_new, m_new

    return None, None
```

```python
def intra_tradet2n(gal, gbl, ml, parameter, matrix, size, f1, f2):

    idx1, idx2 = encounter1(matrix)

    ga1 = gal[idx1]; ga2 = gal[idx2]
    gb1 = gbl[idx1]; gb2 = gbl[idx2]
    m1  = ml[idx1];  m2  = ml[idx2]

    parameter1 = parameter[idx1]
    parameter2=parameter[idx2]
    ledge = parameter1[-1][0]
    hedge = parameter1[-1][1]

    alpha = parameter1[0]
    eta   = parameter1[1]

    M  = m1 + m2
    Ga = ga1 + ga2
    Gb = gb1 + gb2

    i = 0
    while i < 50:

        # ---------- proposal ----------
        new_ga1 = random_pick(ga1, Ga, 0.3)
        new_m1  = random_pick(m1,  M,  0.3)
        new_gb1 = random_pick(gb1, Gb, 0.3)

        dga1 = new_ga1 - ga1
        dm   = new_m1  - m1

        # ---------- price admissibility ----------
        denom = (ga1 - new_ga1)
        if abs(denom) < 1e-12:
            i += 1
            continue

        price = (new_m1 - m1) / denom

        if not (ledge <= price <= hedge):
            i += 1
            continue

        # ---------- conditional weights ----------
        # w(g',m' | g,m) = (g+g')^(alpha-1) (m+m')^(eta-1)

        w1_old = (ga1 + ga1)**(alpha-1) * (m1 + m1)**(eta-1)
        w1_new = (ga1 + new_ga1)**(alpha-1) * (m1 + new_m1)**(eta-1)

        new_ga2 = Ga - new_ga1
        new_m2  = M  - new_m1

        w2_old = (ga2 + ga2)**(alpha-1) * (m2 + m2)**(eta-1)
        w2_new = (ga2 + new_ga2)**(alpha-1) * (m2 + new_m2)**(eta-1)

        w_old = w1_old * w2_old
        w_new = w1_new * w2_new

        # ---------- Metropolis ----------
        accept = False
        if w_new >= w_old:
            accept = True
        else:
            if random.random() < (w_new / w_old):
                accept = True

        if accept:
            ga1 = new_ga1
            m1  = new_m1
            gb1 = new_gb1

            ga2 = Ga - ga1
            m2  = M  - m1
            gb2 = Gb - gb1

        i += 1

    # ---------- relax back ----------
    gal[idx1] += size * (ga1 - gal[idx1])
    gbl[idx1] += size * (gb1 - gbl[idx1])
    ml[idx1]  += size * (m1  - ml[idx1])

    gal[idx2] += size * (ga2 - gal[idx2])
    gbl[idx2] += size * (gb2 - gbl[idx2])
    ml[idx2]  += size * (m2  - ml[idx2])

    return gal, gbl, ml

```

```python
import random

def get_neighbors(lst, index, k=1):
    n = len(lst)
    neighbors = []

    for i in range(index - k, index + k + 1):
        if i == index:
            continue  # Skip the center element
        neighbors.append(lst[i % n])

    return neighbors
```

```python
def get_neighbors2(lst, index, rows=100, cols=10):
    
    n = len(lst)
    
    
   
    row = index // cols
    col = index % cols

    
    up_row = (row - 1) % rows
    down_row = (row + 1) % rows

   
    up_index = up_row * cols + col
    down_index = down_row * cols + col

    return [lst[up_index], lst[down_index]]
```

```python
def get_trade_info(lst, index):
    n = len(lst)
    if not (0 <= index < n):
        raise IndexError("Index out of range")

    # Step 1: Calculate immediate neighbors (Left/Right in a circle)
    idx_prev = (index - 1) % n
    idx_next = (index + 1) % n

    # Step 2: Calculate neighbors' neighbors (Two steps away)
    idx_prev_prev = (idx_prev - 1) % n
    idx_next_next = (idx_next + 1) % n

    # Step 3: Grouping
    # n1: The "previous" side chain
    n1 = [idx_prev, idx_prev_prev]
    # n2: The "next" side chain
    n2 = [idx_next, idx_next_next]
    
    # n3: All unique excluded neighbors
    excluded_set = {idx_prev, idx_prev_prev, idx_next, idx_next_next}
    n3 = list(excluded_set)

    # Step 4: Generate tradable list
    # Exclude self and the 4 neighbors in n3
    all_excluded = excluded_set | {index}
    tradable_list = [i for i in range(n) if i not in all_excluded]

    return n1, n2, n3, tradable_list
```

```python
import random

def encounter5(lst):
    """
    Select two indices (idx1, idx2) such that:
    - idx1 is random
    - idx2 is NOT:
        - idx1 itself
        - idx1's neighbor
        - idx1's neighbor's neighbor
    """

    n = len(lst)

    # Step 1: pick idx1 freely
    idx1 = random.randrange(n)

    # Step 2: compute forbidden set for idx1
    idx_prev = (idx1 - 1) % n
    idx_next = (idx1 + 1) % n
    idx_prev_prev = (idx_prev - 1) % n
    idx_next_next = (idx_next + 1) % n

    forbidden = {idx1, idx_prev, idx_next, idx_prev_prev, idx_next_next}

    # Step 3: admissible trading partners
    candidates = [i for i in range(n) if i not in forbidden]

    # Safety check (for very small n)
    if not candidates:
        raise ValueError("No valid trading partner available")

    idx2 = random.choice(candidates)

    return idx1, idx2
```

```python
def get_neighbors3(lst, index):
    """
    Returns a 5-element list centered at index:
    [left-left, left, center, right, right-right]
    with circular wrap-around.
    """
    n = len(lst)
    
    # Calculate indices using modulo for ring logic
    # -2, -1, 0, +1, +2 relative to index
    idx_ll = (index - 2) % n
    idx_l  = (index - 1) % n
    idx_c  = index         # center
    idx_r  = (index + 1) % n
    idx_rr = (index + 2) % n
    
    # Return values in order
    return [lst[idx_ll], lst[idx_l], lst[idx_c], lst[idx_r], lst[idx_rr]]
```

```python
def pm(ml1, p1):
    # Check if the index is valid
   
    
    # Pick the item at idx1
  
    
    # Initialize the product with 1
    product = 1
    
    # Multiply all items except the one at idx1
    for i, item in enumerate(ml1):
            
            product *= item**p1
    
    return  product
```
